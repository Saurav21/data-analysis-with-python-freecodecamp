```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("medical_examination.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
weight = data["weight"]

weight.head()
```

```python
height = data["height"] / 100

height**2
```

```python
bmis= round(weight / height**2,2)
bmis.head()
```

```python
data["overweight"] = bmis.apply(lambda x : 1 if x>25 else 0)
    
```

```python
data
```

```python
data["cholesterol"] = data["cholesterol"].apply(lambda x: 0 if x == 1 else 0)
```

```python
data
```

```python
data["gluc"] = data["gluc"].apply(lambda x:0 if x == 1 else 0)
```

```python
data
```

```python
data_cat = pd.melt(data, id_vars = ["cardio"], value_vars = ["cholesterol", "gluc", "smoke", "alco", "active","overweight"])
```

```python
data_cat["total"] = 1
```

```python
data_cat = data_cat.groupby(["cardio", "variable","value"], as_index = False).count()
```

```python
data_cat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.catplot(x = "variable", y = "total", data = data_cat, hue = "value", kind = "bar", col = "cardio")
plt.show()
```

```python
heat = data[(data['ap_lo'] <= data['ap_hi']) &
(data['height'] >= data['height'].quantile(0.025)) &
(data['height'] <= data['height'].quantile(0.975)) &
(data['weight'] >= data['weight'].quantile(0.025)) &
(data['weight'] <= data['weight'].quantile(0.975))] 
```

```python
corr = heat.corr(method = "pearson")
```

```python
mask = np.triu(corr)
```

```python
plt.figure(figsize = (10,12))
sns.heatmap(corr, annot = True, mask = mask)
```
